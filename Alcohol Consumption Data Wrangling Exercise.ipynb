```python
# Import working tools

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Import drinks dataset

```python
drinksdf = pd.read_csv('drinks.csv')
drinksdf
```

# Convert object to numeric data type

```python
# Convert values to numeric
drinksdf['beer_servings'] = pd.to_numeric(drinksdf['beer_servings'], errors='coerce')
drinksdf['spirit_servings'] = pd.to_numeric(drinksdf['spirit_servings'], errors='coerce')
drinksdf['wine_servings'] = pd.to_numeric(drinksdf['wine_servings'], errors='coerce')
```

# Fill missing values

```python
drinksdf['beer_servings'] = drinksdf['beer_servings'].fillna(drinksdf['beer_servings'].mean())
drinksdf['spirit_servings'] = drinksdf['spirit_servings'].fillna(drinksdf['spirit_servings'].mean())
drinksdf['wine_servings'] = drinksdf['wine_servings'].fillna(drinksdf['wine_servings'].mean())
```

```python
# Convert Values to int
drinksdf['beer_servings'] = drinksdf['beer_servings'].astype(int)
drinksdf['spirit_servings'] = drinksdf['spirit_servings'].astype(int)
drinksdf['wine_servings'] = drinksdf['wine_servings'].astype(int)
```
