```python
# Import working tools

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import drinks dataset

drinksdf = pd.read_csv('drinks.csv')
drinksdf
```

### Clean data of Missing Values

```python
# Get rid of whitespaces
drinksdf['beer_servings']=drinksdf['beer_servings'].str.strip('?').replace('','NaN')
drinksdf['spirit_servings']=drinksdf['spirit_servings'].str.strip('?').replace('','NaN')
drinksdf['wine_servings']=drinksdf['wine_servings'].str.strip('?').replace('','NaN')
```

```python
# # Fill missing values
drinksdf[drinksdf['beer_servings']== 'NaN'] = 0
drinksdf[drinksdf['spirit_servings']== 'NaN'] = 0
drinksdf[drinksdf['wine_servings']== 'NaN'] = 0
# This was done this way because the columns only contained '?' symbols and at this stage i couldn't handle it
```

```python
# Convert Values to int
drinksdf['beer_servings'] = drinksdf['beer_servings'].astype(int)
drinksdf['spirit_servings'] = drinksdf['spirit_servings'].astype(int)
drinksdf['wine_servings'] = drinksdf['wine_servings'].astype(int)
```

```python
# 34 Ounces = 1litre

# Average-by-voume alcohol in drinks:
#     Spirits = 37%
#     Wine = 11.6%
#     Beer = 4.5%

# Source: www.livescience.com%2F32735-how-much-alcohol-is-in-my-drink
```

```python
abv_spirit = 37/100
abv_wine = 11.6/100
abv_beer = 4.5/100
```

```python
# Take each serving as 1bottle

drinksdf['total_litres_of_pure_alcohol'] = (drinksdf['beer_servings'] * abv_beer) + (drinksdf['spirit_servings']*abv_spirit) + (drinksdf['wine_servings']*abv_wine)
drinksdf
```

```python
# Export dataset
drinksdf.to_csv('alcohol-consumption-data.csv')
```
